```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xs = np.arange(-0.1, 1.1, 0.001)
ys1 = np.clip(xs, 0, 1)
plt.plot(xs, ys1)

ys2 = np.clip(np.sqrt(np.clip(xs, 0, 1)), 0, 1)
plt.plot(xs, ys2)

ys3 = np.clip(xs ** 3, 0, 1)
plt.plot(xs, ys3)

plt.title('Probability Functions for Collection')
plt.xlabel('$\\delta / \\max(\\delta_i)$')
plt.ylabel('Collection Probability')
plt.legend(['$y=x$', '$y=\\sqrt{x}$', '$y=x^3$'])

plt.savefig('ProbabilityFunctions.png')
plt.show()
```
